```python
import pandas as pd
```

```python
apple=pd.read_csv('D:/Data/aapl.csv',index_col=[0],parse_dates=[0])
amazon=pd.read_csv('D:/Data/amzn.csv',index_col=[0],parse_dates=[0])
fb=pd.read_csv('D:/Data/fb.csv',index_col=[0],parse_dates=[0])
google=pd.read_csv('D:/Data/goog.csv',index_col=[0],parse_dates=[0])
netflix=pd.read_csv('D:/Data/nflx.csv',index_col=[0],parse_dates=[0])
```

```python
apple['ticker']='aapl'
amazon['ticker']='amzn'
fb['ticker']='fb'
google['ticker']='goog'
netflix['ticker']='nflx'
```

```python
faang=pd.concat([apple,amazon,fb,google,netflix],axis=0, join="inner").sort_index()
faang
```

```python
faang.to_csv('D:/Data/bruh.csv')
```

# EX2

```python
data = pd.read_csv('D:/Data/covid19_cases.csv')
data
```

```python
data['date']=data.dateRep
data
```

```python
data.set_index('date', inplace = True)
```

```python
data.sort_index()
```

```python
data.replace(to_replace=('United_States_of_America'), value=('USA'), inplace=True)
data.replace(to_replace=('United_Kingdom'), value=('UK'), inplace=True)
data
```

```python
data.sort_values(by=['countriesAndTerritories'], ascending= True)
```

```python
cleaned =data[data.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
cleaned
```

```python
pdf=data.pivot(columns='countriesAndTerritories',values='cases').fillna(0)
pdf
```
